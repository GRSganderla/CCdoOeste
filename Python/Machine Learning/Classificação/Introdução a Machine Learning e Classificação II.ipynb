## Importando dados e definindo o conjunto treino e teste

```python
import pandas as pd

data = pd.read_csv('tracking.csv')
data
```

```python
colunas = {'home' : 'principal', 'how_it_works': 'como_funciona', 'contact': 'contato', 'bought' : 'comprou'}
data = data.rename(columns = colunas)
```

```python
x = data[['principal', 'como_funciona', 'contato']]
y = data['comprou']
```

```python
treino_x = x[:75]
treino_y = y[:75]
teste_x  = x[75:]
teste_y  = y[75:]
print(f'Treino com {treino_x.shape[0]} elementos e teste com {teste_x.shape[0]} elementos')
```

## Criando modelo

```python
from sklearn.svm import LinearSVC as lsvc
from sklearn.metrics import accuracy_score as accuracy

modelo = lsvc()
modelo.fit(treino_x, treino_y)

prev = modelo.predict(teste_x)
acuracia = accuracy(teste_y, prev)
print(f'A Acuracia foi de {acuracia*100:.2f}')
```

## Modo diferente pronto de separa conjunto de dados

```python
from sklearn.svm import LinearSVC as lsvc
from sklearn.metrics import accuracy_score as accuracy
from sklearn.model_selection import train_test_split as tt_split

SEED = 20

treino_x, teste_x, treino_y, teste_y = tt_split(x, y, random_state = SEED, test_size = .25, stratify = y)
print(f'Treino com {treino_x.shape[0]} elementos e teste com {teste_x.shape[0]} elementos')
modelo = lsvc()
modelo.fit(treino_x, treino_y)

prev = modelo.predict(teste_x)
acuracia = accuracy(teste_y, prev)
print(f'A Acuracia foi de {acuracia*100:.2f}')
```
