```python
import pandas as pd

filmes = pd.read_csv('data/movies.csv')
filmes.head()
```

```python
trocar = {'movieId':'filmeId', 'title':'titulo','genres':'generos'}
filmes.rename(columns = trocar, inplace = True)
filmes = filmes.set_index('filmeId')
filmes.head()
```

```python
notas = pd.read_csv('data/ratings.csv')
trocar = {'userId':'usuarioId', 'movieId':'filmeId', 'rating':'nota', 'timestamp':'momento'}
notas.rename(columns = trocar, inplace = True)
notas.head()
```

# Primeira tentativa de recomendação

```python
total_de_votos = notas['filmeId'].value_counts()
total_de_votos.head()
```

```python
filmes.loc[318]
```

```python
filmes['total_de_votos'] = total_de_votos
filmes.head()
```

```python
filmes.sort_values('total_de_votos', ascending = False)
```

```python
notas_medias = notas.groupby('filmeId').mean()['nota']
notas_medias.head()
```

```python
filmes['nota_media'] = notas_medias
filmes.head()
```

```python
filmes.sort_values('total_de_votos', ascending = False).head()
```

# Uma segunda Heurística

```python
filmes.sort_values('nota_media', ascending = False).head()
```

```python
filmes.query('total_de_votos >= 10').sort_values('nota_media', ascending = False).head()
```

```python
filmes_com_50mais_votos = filmes.query('total_de_votos >= 50')
filmes_com_50mais_votos.sort_values('nota_media', ascending = False).head()
```
