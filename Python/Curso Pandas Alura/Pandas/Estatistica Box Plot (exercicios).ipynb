```python
import pandas as pd
data = pd.read_csv('data/aluguel_amostra.csv', sep =';')
data
```

```python
Q1 = data['Valor m2'].quantile(.25)
Q3 = data['Valor m2'].quantile(.75)
IIQ = Q3-Q1
lim_inf = Q1 - 1.5*IIQ
lim_sup = Q3 + 1.5*IIQ
print(f'|Q1| -> {Q1}\n|Q3| -> {Q3}\n|IIQ| -> {IIQ}\n|Q1 - 1.5*IIQ| -> {lim_inf}\n|Q3 + 1.5*IIQ| -> {lim_sup}\n')
```
