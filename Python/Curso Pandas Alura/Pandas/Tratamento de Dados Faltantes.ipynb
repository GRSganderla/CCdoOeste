## Tratamento de Dados Faltantes (Cont.)

```python
selecao = (data['Tipo'] == 'Apartamento') & (data['Condominio'].isnull())
```

```python
A = data.shape[0]
data = data[~selecao]
B = data.shape[0]
A - B
```

```python
data[data['Condominio'].isnull()].shape[0]
```

```python
data = data.fillna({'Condominio': 0, 'IPTU':0})
```
