## Criando Agrupamentos

```python
import pandas as pd
data = pd.read_csv('data/aluguel_residencial.csv', sep=';')
```

```python
data.head(10)
```

```python
bairros = ['Barra da Tijuca', 'Copacabana', 'Ipanema', 'Leblon', 'Botafogo', 'Flamengo', 'Tijuca']
selecao = data['Bairro'].isin(bairros)
selecao
```

```python
data = data[selecao]
data
```

```python
data['Bairro'].drop_duplicates()
```

```python
grupo_bairros = data.groupby('Bairro')
type(grupo_bairros)
```

```python
grupo_bairros.groups
```

```python
for bairro, dado in grupo_bairros:
    print(bairro)
```

```python
for bairro, dado in grupo_bairros:
    print(dado)
```

```python
for bairro, dado in grupo_bairros:
    print(f'{bairro} -> {dado.Valor.mean()}')
```

```python
grupo_bairros[['Valor', 'Condominio']].mean().round(2)
```

## Estatisticas Descritivas

```python
grupo_bairros['Valor'].describe().round(2)
```

```python
grupo_bairros['Valor'].aggregate(['min', 'max', 'sum']).rename(columns = {'min' : 'Minímo', 'max':'Maxímo', 'sum': 'Soma'})
```

```python
%matplotlib inline 
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (20, 10))
```

```python
fig = grupo_bairros['Valor'].mean().plot.bar(color = 'blue')
fig.set_ylabel('Valor do Aluguel')
fig.set_title('Valor Medio do Aluguel por Bairro', {'fontsize': 22})
```

```python
fig = grupo_bairros['Valor'].max().plot.bar(color = 'blue')
fig.set_ylabel('Valor do Aluguel')
fig.set_title('Valor Medio do Aluguel por Bairro', {'fontsize': 22})
```
